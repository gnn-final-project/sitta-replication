```python
import os
import subprocess

# âœ… ì„¤ì •
source_folder = "../../data/processed/plant pathology/healthy"
output_folder = "../../data/processed/plant pathology/sick_singan"
model_dir = "./results/healthy2sick"
os.makedirs(output_folder, exist_ok=True)

# âœ… ëª¨ë“  healthy ì´ë¯¸ì§€ì— ëŒ€í•´ SinGAN ë³€í™˜
image_list = sorted(os.listdir(source_folder))

for i, fname in enumerate(image_list):
    input_path = os.path.join(source_folder, fname)
    save_name = f"sick_{i+1:03d}.jpg"
    output_path = os.path.join(output_folder, save_name)

    print(f"ğŸ“Œ Generating: {save_name}")

    # ìƒì„± ëª…ë ¹
    subprocess.run([
        "python", "main.py",
        "--root", input_path,
        "--evaluation",
        "--model-to-load", os.path.join(model_dir, "g_multivanilla.pt"),
        "--amps-to-load", os.path.join(model_dir, "amps.pt"),
        "--save", "temp_gen",
        "--results-dir", "./results",
        "--num-steps", "10"
    ])

    # ìƒì„±ëœ ì´ë¯¸ì§€ ë³µì‚¬ (s9_sampled.pngê°€ ìµœì¢… ê²°ê³¼)
    gen_path = "./results/temp_gen/s11/s9_sampled.png"
    if os.path.exists(gen_path):
        os.rename(gen_path, output_path)
```
